```python
from google import genai
from google.colab import userdata

GOOGLE_API_KEY = gemini_api_key
client = genai.Client(api_key=GOOGLE_API_KEY)

def generate_ai_news(user):
    prompt = f"Crie uma mensagem para {user['name']} sobre a importância dos investimentos (máximo de 100 caracteres)"

    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=prompt,
        config={
            'system_instruction': 'Você é um especialista em marketing bancário.'
        }
    )

    return response.text.strip()

for user in users:
    print(f"Gerando mensagem para {user['name']}...")
    mensagem_gerada = generate_ai_news(user)
    user['news'].append(mensagem_gerada)

texto = "Processo concluído! Resultado final:"
largura_menu = 60

print(f"\n{texto.center(largura_menu, '-')}\n")

import pprint
pprint.pprint(users)
```

```python
df = pd.DataFrame(users)

df['news'] = df['news'].apply(lambda x: x[0] if x else "")

df.to_excel('relatorio_marketing_gemini.xlsx', index=False)
df.to_csv('relatorio_marketing_gemini.csv', index=False, encoding='utf-8-sig')

print("Load concluído! Arquivos salvos com sucesso.")
```
